```python
import numpy as np 
import matplotlib.pyplot as plt
import json 
import pandas as pd 
import os
```

```python
data_directory = 'bp_05_05_results'
sparities = ['005', '010'] 


counts = {}
accuracies = {}
for sparity in sparities:
    data_dir = os.path.join(data_directory, sparity)
    accuracies[sparity] = []
    for file in os.listdir(data_dir):
        with open(os.path.join(data_dir, file), 'r') as f:
            data = json.load(f)
            if data['parameters']['K'] == 4: 
                accuracy = data['results']['PairSamplingObservation']['stats']['accuracy']
                accuracies[sparity].append(accuracy)
        counts[sparity] = counts.get(sparity, 0) + 1
        
# print out accuaries
for sparity in sparities:
    print(f"Sparity: {sparity}")
    print(f"Number of files: {counts[sparity]}")
    print(f"Accuracies: {accuracies[sparity]}")
    print(f"Mean accuracy: {np.mean(accuracies[sparity])}")
    print(f"Std accuracy: {np.std(accuracies[sparity])}")
    print()
```
